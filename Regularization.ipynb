```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
```

```python
df = pd.read_csv("/content/regression_training_dataset1_5537be90-411f-4a6b-8ea4-7bb197181da5_268917_.csv")
```

```python
df.info()
```

```python
x = df[["StudyHoursPerWeek", "AttendancePercent", "AssignmentsCompleted", "ExperienceYears"]]
y=df["PerformanceScore"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state = 42)
```

```python
scaler = StandardScaler()
x_train_scaler = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
ols = LinearRegression()
ols.fit(x_train, y_train)
```

```python
model_ridge = Ridge(alpha=0.5)
model_ridge.fit(x_train, y_train)
```

```python
model_lasso = Lasso(alpha=5.0)
model_lasso.fit(x_train, y_train)
```

```python
model_enet = ElasticNet(alpha=1, l1_ratio=0.5)
model_enet.fit(x_train, y_train)
```

```python
ols = LinearRegression()
ridge = Ridge(alpha=5.0)
lasso = Lasso(alpha=0.5)
elastic = ElasticNet(alpha=1, l1_ratio=0.5)

ols.fit(x_train, y_train)
ridge.fit(x_train, y_train)
lasso.fit(x_train, y_train)
elastic.fit(x_train, y_train)
```

```python
print("OLS Coefficients:", ols.coef_)
print("Ridge Coefficients:", ridge.coef_)
print("Lasso Coefficients:", lasso.coef_)
print("ElasticNet Coefficients:", elastic.coef_)
```

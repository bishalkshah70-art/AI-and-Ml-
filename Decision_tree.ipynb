```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, accuracy_score
```

```python
df = pd.read_csv("/content/customer_segmentation1_81858f5b-5485-404f-9a38-518bcb69ee52_268917_.csv")
df = pd.read_csv("/content/online_shoppers_intention_785c2612-e1e8-42f0-8e41-e1024644bb47_268917_.csv")
```

```python
x_raw = df.drop(columns=["Revenue"])
y =df["Revenue"].astype(int)
```

```python
x_encoded = pd.get_dummies(x_raw, columns=['Month', 'VisitorType'], drop_first=True)
```

```python
TRANING_FEATURE_SCHEMA = x_encoded.columns.tolist()
```

```python
x_train,x_test,y_train,y_test = train_test_split(
    x_encoded,y,test_size=0.2,random_state=42,stratify=y
)
```

```python
print('Step 2; Building Regularizaed Decision Tree Structure...')
clf_tree = DecisionTreeClassifier(criterion='gini', max_depth=4, random_state=42)
clf_tree.fit(x_train,y_train)
```

```python
print("\n===SYSTEM EVALUATION PERFORMANCE DASHBOARD===")
y_pred = clf_tree.predict(x_test)


print(f"Operational Model Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(classification_report(y_test, y_pred, target_names=['Abandoned Cart', 'ConvertedPurchase']))
```

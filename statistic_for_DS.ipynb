```python
import numpy as np
from scipy import stats
```

```python
rent_prices = np.array([1200, 1350, 1200, 1400, 1550, 1300,1200, 12000])
```

```python
# calculating central tedenncy
mean_rent=np.mean(rent_prices)
median_rent=np.median(rent_prices)
```

```python
mode_result=stats.mode(rent_prices, keepdims=True)
print(mode_result)
mode_rent=mode_result.mode[0]
mode_count=mode_result.count[0]
```

```python
print(f"Mean Rent: ${mean_rent}")
print(f"Median Rent: ${median_rent}")
print(f"Mode Rent: ${mode_rent}")
```

```python
# Dataset Daily tempratures for two cities over 5 day(Both have a mean of 25c)
city_a_temps=np.array([24,25,26,25,25])
city_b_temps=np.array([10,40,5,45,25])

var_a=np.var(city_a_temps)
var_b=np.var(city_b_temps)
```

```python
std_a=np.std(city_a_temps)
std_b=np.std(city_b_temps)
print(f"Variance in City A: {var_a: 2f}")
print(f"Variance in City B: {var_b: 2f}")
print(f"Standard Deviation in City A: {std_a: 2f}")
print(f"Standard Deviation in City B: {std_b: 2f}")
```

```python
# vraiables: making spend ($), Online Sales($) , and store complaints(count)
marketing_spend = np.array([10,20,30,40,50])
online_sales = np.array([25, 45, 55, 75, 95])
complaints = np.array([15,12,10,4,2])
```

```python
cov_matrix = np.cov(marketing_spend, online_sales)
cov_matrix1 = np.cov(marketing_spend, complaints)
print(cov_matrix)
print(cov_matrix1)
```

```python
corr_matrix_positive=np.corrcoef(marketing_spend, online_sales)
corr_matrix_negative=np.corrcoef(marketing_spend, complaints)
print(corr_matrix_positive)
print(corr_matrix_negative)
```

```python
correlation_spend_sales = np.corrcoef(marketing_spend, online_sales)[0, 1]
correlation_spend_complaints = np.corrcoef(marketing_spend, complaints)[0, 1]

print(f"Correlation between Marketing Spend and Online Sales: {correlation_spend_sales}")
print(f"Correlation between Marketing Spend and Complaints: {correlation_spend_complaints}")
```

```python
import numpy  as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load dataset
df=pd.read_csv("/content/automation_projects.csv")
```

```python
print("="*50)
print("DATASET OVERVIEW")
print("="*50)

print("Rows and Columns:", df.shape)

print("\nColumn Names:")
print(df.columns.tolist())

print("\nData Types:")
print(df.dtypes)

print("\nFirst 5 Rows:")
display(df.head())
```

```python
#Data Cleaning
print("\nMissing Values")
print(df.isnull().sum())
```

```python
#remove duplicates
df=df.drop_duplicates()
```

```python
print("\nDataset shape After Cleaning:",df.shape)
```

```python
#Descriptive statisctics
print("\nSummary Statistics")
df.describe()
```

```python
# numeric columns analysis
num_cols = df.select_dtypes(include=np.number).columns

for col in num_cols:
    print(f"\n----- {col} -----")
    print("Mean:", df[col].mean())
    print("Median:", df[col].median())
    print("Maximum:", df[col].max())
    print("Minimum:", df[col].min())
    print("Standard Deviation:", df[col].std())
```

```python
#Top Industries
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    print(f"\nTop Values in {col}")
    print(df[col].value_counts().head())
```

```python
if len(num_cols) > 1:

    corr = df[num_cols].corr()

    plt.figure(figsize=(10,8))
    plt.imshow(corr, cmap="coolwarm")
    plt.colorbar()
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
    plt.yticks(range(len(corr.columns)), corr.columns)

    plt.show()
```

```python
for col in num_cols:
    plt.figure(figsize=(6,4))
    plt.hist(df[col].dropna(), bins=20)

    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
for col in num_cols:
    plt.figure(figsize=(6,4))
    plt.hist(df[col].dropna(), bins=20)

    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
#EDA visulation
plt.figure(figsize=(8,4))
plt.hist(df["roi_percent"],bins=20)
plt.title("ROI_Distribution")
plt.xlabel("ROI %")
plt.ylabel("ROI_Distribution")
plt.show()
```

```python
if "roi_percent" in df.columns:

    print("\nROI ANALYSIS")
    print("Average ROI:", round(df["roi_percent"].mean(),2))
    print("Highest ROI:", round(df["roi_percent"].max(),2))
    print("Lowest ROI:", round(df["roi_percent"].min(),2))
```

```python
df["project_status"].value_counts().plot(kind="pie",autopct="%1.1f%%")
plt.ylabel("")
plt.show()
```

```python
print("\n===== FINAL REPORT =====")
print("Total Records:", len(df))
print("Total Features:", len(df.columns))
print("Numeric Columns:", len(num_cols))
print("Categorical Columns:", len(cat_cols))
print("Missing Values:", df.isnull().sum().sum())
print("Duplicate Rows Removed Successfully")

# Note: missing values are concentrated in 'completion_date' and only occur
# for projects with project_status == 'Active' (i.e. not finished yet),
# so this is expected/structural, not bad data.
print("\nMissing completion_date by project_status:")
print(df.groupby("project_status")["completion_date"].apply(lambda s: s.isnull().sum()))
```

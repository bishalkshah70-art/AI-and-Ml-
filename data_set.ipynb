```python
import pandas as pd
import numpy as np
```

```python
df_sales = pd.read_csv("/content/advanced_pandas_sales_dataset (1).csv")
```

```python
df_sales.info()
```

```python
print(df_sales.head)
```

```python
category_revenue = df_sales.groupby('Category')['Revenue'].sum().reset_index
```

```python
print("--------Tortal Revenue by Category---------")
print(category_revenue)
```

```python
region_matrix = df_sales.groupby("Region").agg(
    Total_Units_sold=("Quantity","sum"),
    Average_Item_price=("UnitPrice","mean"),

).reset_index()
```

```python
print(region_matrix )
```

```python
manager_data={
      "Region" :["East", "West", "North", "South"],
      "Regional_Director" :["Alice Cooper", "BOb Marley", "Charlie Chaplin", "David Beckham"]

  }
df_manager = pd.DataFrame(manager_data)
```

```python
# merger according to database using "Region" as th link key
df_enriched = pd.merge(df_sales, df_manager, on="Region", how="left")
print("---------Enriched Data---------")
print(df_enriched[["OrderID", "Region", "Category", "Revenue", "Regional_Director"]].head(3))
```

```python
revenue_pivot=pd.pivot_table(
    df_sales,
    values="Revenue",
    index="Region",
    columns="Category",
    aggfunc="sum"
)
print("---------Revenue vs Category---------")
print(revenue_pivot)
```

```python
order_counts_crosstab = pd.crosstab(
    index=df_sales["Region"],
    columns=df_sales["Category"],
    values=df_sales["OrderID"],
    aggfunc="count",
    margins=True,
    margins_name="Total_Orders"
)

print("\n----Order frequency crosstab(Transaction counts)----")
print()
print(order_counts_crosstab)
```

```python
print(f"Original OrderDate column data type: {df_sales['OrderDate'].dtype}")
```

```python
df_sales['OrderDate'] = pd.to_datetime(df_sales['OrderDate'])
```

```python
print(f"Modified OrderDate column data type: {df_sales['OrderDate'].dtype}")
```

```python
df_sales['Month_Name'] = df_sales['OrderDate'].dt.strftime("%B")
df_sales['Day_of_Week'] = df_sales['OrderDate'].dt.day_name()
```

```python
monthly_trends = (
    df_sales
    .groupby(df_sales["OrderDate"].dt.to_period("M"))["Revenue"]
    .sum()
    .reset_index()
)

monthly_trends["OrderDate"] = monthly_trends["OrderDate"].dt.to_timestamp()
```

```python
print(monthly_trends)
```

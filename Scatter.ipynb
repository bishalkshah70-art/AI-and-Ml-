```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0,10,100)
print(x)
```

```python
y1 = x**2 # quadratic growth func: f(x)=x^2
y2 = 10*x # linear baseline growth: g(x) = 10x

plt.plot(x,y1,label="f(x)=x^2", color="blue", linestyle="-",linewidth=2)
plt.plot(x,y2,label="g(x)=10x", color="red", linestyle="-",linewidth=2)

plt.title("Line Plot",fontsize=13,fontweight="bold")
plt.xlabel("Independent variable(x-axis)",fontsize=12)
plt.ylabel("Computed output(y-axis)",fontsize=12)
plt.legend(loc="center")
plt.grid(True,linestyle=":",alpha=0.8)
plt.show()
plt.savefig("line_plot.png")
plt.close()



```

```python
df_election = pd.read_csv("/content/election_data.csv")
df_district = pd.read_csv("/content/district_data (1).csv")
df_province = pd.read_csv("/content/province_data.csv")
```

```python
df_master = pd.merge(df_election,df_district,on="district_id", how ="inner").merge(df_province,on="province_id",how ="left")
```

```python
plt.scatter(df_master["age"],df_master["votes"],color="#00FFFF",alpha=0.5)
plt.title("Scatter Plot",fontsize=13,fontweight="bold")
plt.xlabel("Candidate Age(years)",fontsize=12)
plt.ylabel("Total Vots Counted",fontsize=12)
plt.grid(True,alpha=0.3)
plt.savefig("scatter_plot.png")
plt.close()
```

```python
sns.scatterplot(data=df_master,x="age",y="votes")
```

```python
party_count=df_master["party"].value_counts().head(5).sort_values(ascending=True)
print(party_count)
```

```python
plt.bar(party_count.index,party_count.values,color="#00FFFF", edgecolor="black")
plt.title("Top 5 political Party")
plt.xlabel("Registered Candidates")
plt.ylabel("Political Party")
plt.xticks(rotation="vertical")

plt.savefig("bar_plot.png")
```

```python
gender_distribution=df_master["gender"].value_counts()
```

```python
gender_distribution
```

```python
plt.pie(
    gender_distribution.values,
    labels=gender_distribution.index,
    autopct="%1.1f",
    colors=["#00FFFF", "lightpink", "Black"],
    startangle=140,
    pctdistance=0.9,
    explode=[0,0,0.1]

)
plt.title("Gender Distribution")
plt.savefig("pie_chart.png")
```

```python
plt.hist(df_master["age"],bins=15,color="Orange", edgecolor="black",alpha=0.8)
plt.title("National candidateAge Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True,axis="y", linestyle="--",fontsiz=12)
plt.savefig("histogram.png")
```

```python
province_order=df_master.groupby("province_name")["age"].median().sort_values().index
print(province_order)
```

```python
sns.set_theme(style="whitegrid")
sns.boxplot(data=df_master,x="province_name",y="age",order=province_order,palette="Set3")
plt.xticks(rotation="horizontal")
plt.title("candidate Age Distribution")
plt.xlabel("Candidate Age")
plt.ylabel("Province")
plt.show()
plt.savefig("set_theme.png")
```
